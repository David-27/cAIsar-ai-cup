```python
import pandas as pd
import matplotlib.pyplot as plt
from typing import List, Set, Dict, Tuple, Optional

import sys, os
sys.path.append(os.path.dirname(os.path.abspath('')).split("src")[0])

from utils.value_smoothing import smooth
path = "C:\Informatik\workspaces\workspace_python\AI-Cup\logs\csv-03222022.log"
df = pd.read_csv (path)
```

```python
values :List[float] = df["val_accuracy"].tolist()
smooth_constant : float = 0.6
```

```python
values_x = range(0,len(values))
values_y = values
print("Max value raw",round(max(values_y),4))

plt.plot(values_x, values_y,label="Raw")

values_x = range(0,len(values))
values_y = smooth(values_y,smooth_constant)
print("Max value smoothed",round(max(values_y),4))
plt.plot(values_x, values_y,label="Smoothed")
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
```

```python
# clear figure
plt.clf

for column_name in df:

    plt.clf

    legend : str = column_name
    values = df[column_name].tolist()
    smoothed_values = smooth(values,smooth_constant)

    values_x = range(0,len(values))

    plt.plot(values_x, values,label="raw")
    plt.plot(values_x, smoothed_values,label="smooth")

    plt.xlabel('Epoch')
    plt.ylabel(legend)
    plt.legend()
    plt.show()
```
